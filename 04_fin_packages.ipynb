## Data

```python
url = 'https://certificate.tpq.io/findata.csv'
```

```python
raw = pd.read_csv(url, index_col=0, parse_dates=True)
```

```python
raw.info()
```

```python
# ticker = 'SPY'
ticker = '.SPX'
ticker_ = 'AAPL.O'
```

```python
data = pd.DataFrame(raw[[ticker, ticker_]]).dropna()
```

```python
data['r'] = data[ticker].pct_change()
```

```python
data['r_'] = data[ticker_].pct_change()
```

```python
r = data['r']
```

```python
r_ = data['r_']
```

## Basics
